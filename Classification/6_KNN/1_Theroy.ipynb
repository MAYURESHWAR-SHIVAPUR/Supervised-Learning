## **Defination**

`It is supervised learning algorithum which is used for both regression and classification `

`It uses distance formula to predict the output`

`It uses 2 distance formula EUCLIDEN, MANHATTON`

`It is a lazy algorithum, during model traning it just stores the data not learns the data.`

`It calculates the result using (Eucliden, Manhatton) when user enter new input(Prediction)`

`K` `How many Neaibours`

`N` `nearest naibour`
