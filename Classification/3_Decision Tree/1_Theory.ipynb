#### **1.1 Random Forest**


`It is supervised learning algorithum which comes under bagging ensomble technique and we can use it for both REGRESSION and CLASSIFICATION`

---

**Working**

**`STEP 1 =>`**  `It creates 100 decission tree internally`

**`STEP 2 =>`**  `It  creates random sample for all 100 trees with the help of boosttraping sampling technique`

**`STEP 3 =>`**  `It parallel trains all 100 Trees so trees are independent`

---

**How RANDOM FOREST predicts the output**

**`Classification`**  `Takes Majority(higest frequcy of the elemet)`

**`Regression`**  `Takes Average`

---

**When not to use Random Forest**

`When decission tree is under fit (because it reduce baise)`

`for largver dataset we should not use RANDOM FOREST (bec the tree depth increase also it take more time and will not get good accuracy)`
