# **Importing and reading files**

```python
import pandas as pd

from xgboost import XGBRegressor, XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("../../../Data/loan.csv")
df
```

# **Extracting columns**

```python
X = df.drop(columns="Loan_Status")
y = df["Loan_Status"]

num_col = X.select_dtypes(include="number").columns
cat_col = X.select_dtypes(include="object").columns
```

# **Converstions**

```python
X[cat_col] = X[cat_col].astype("category")

X[cat_col].info()
```

```python
y = y.map({"Y":1, "N":0})
y.astype("int")
y
```

# **Train test split**

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, train_size=0.8, random_state=42)
ytest
```

# **Model Traning and Predication**

```python
# it handle categorical columns not string so convert the str column to categorical before train_test_spilt)()
model = XGBClassifier(enable_categorical=True, max_depth=25, n_estimators = 150, learning_rate=0.01)
```

```python
model.fit(Xtrain, ytrain)
```

```python
y_train_pred = model.predict(Xtrain)
```

```python
y_test_pred = model.predict(Xtest)
```

# **Model Evaluation**

```python
model.score(Xtrain, ytrain)
```

```python
model.score(Xtest, ytest)
```
