<h2 style="color:#ADFF2F"><b>Precision Score</b></h2>

### **What?**

> it is classification Matrix which tells out of all the predicted values how much percentage values are correctly predicted


### **Syntax**

#### `from sklearn.metrics import precision_score`

#### `precision_score(ActualOutput, Predicted Output, pos_label=1)`

### **Precison Score for positive lable**
 `out of all the positive lables how much percentage lables are correclty prediced as positive`

### **Formula**

`TP/(TP+FP) out of total positives how much is positivly correct(True)`

example 1

> total samples 0(300) 1(700)

|||
|---|---|
|100|200|
|500|200|

`Solution  => 200/(200+200) => 50%`

`Conclusion => out of all the predicted positive lables (400), 50% samples are correctly predicted as positive`

example 2

> total samples spam(400) not spam(100)

|||
|---|---|
|80|20|
|100|300|

`Solution  => 300/(20+300) => 93.75%`

`Conclusion => out of all the predicted spam lables (320), 93.75% lables are correctly predicted as spam`

### **Precison Score for Negative lable** 
`out of all the negative lables how much percentage lables are correclty prediced as negative`

example 1

> total samples 0(300) 1(700)

|||
|---|---|
|100|200|
|500|200|

`Solution  => 100/(100+500) => 16.6%`

`Conclusion => out of all the predicted Negative lables (600), 16.6% samples are correctly predicted as negative`

example 2

> total samples spam(400) not spam(100)

|||
|---|---|
|80|20|
|100|300|

`Solution  => 80/(80+100) => 44.44%`

`Conclusion => out of all the predicted not spam lables (180), 44.44% lables are correctly predicted as not spam`

<h2 style="color:#ADFF2F"><b>F1 Score</b></h2>

### **What?**

`It is a classification metric which balances both precision and recall score and we use this for imbalance Dataset(not equal) It will calculate different accuracy for different category not overall`


### **Formula**

`Neg` `2 x (Precision_Neg x Recall_Neg)/(Precision_Neg + Recall_Neg)`

`Pos` `2 x (Precision_Pos x Recall_Pos)/(Precision_Pos + Recall_Pos)`
