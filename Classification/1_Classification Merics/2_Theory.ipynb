<h1 style="color:#00FFFF"><b>Column Transformer</b></h1>

---
---

## **Defination?**

`It is a ML Technique which combines multiple pre-processing technique and executes parallely`

`In column Transformer we use only pre-processing techinque not an algorithum`

## **Syntax**

`from sklearn.compose import ColumnTransformers`

`preprocessing = ColumnTransformers(transformers = [("name of Preprocessing",preprocessing,columnName(1D)) remainders="drop"||"passthrough"])`

`Column name : must be list of values`

`Remainder : it tells what should be done which remaning columns for dataset() drop=> drops the remaning columns where as passthrought send back same column without any transformation`

## **Example**

`open up`

preprocessing = ColumnTransformers(
    transformers = [
        ("minmax_scale",MinmaxScaler(),["F1","F2"]),
        ("Rubust_scale",RobustScaler(),["F3","F4"]),
        ("encoder",TargetEncoder(),["F5","F6"]),
        ]
    )

new_Xtrain = preprocessing.fit_transform(Xtrain)

`It combines multiple steps like pre-processing and algorithum `

`It executes the steps one by one`

`Use this when we have only pre-processing technique`
