<h1 style="color:skyblue"><b>Importing<b></h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from category_encoders import TargetEncoder
```

<h1 style="color:skyblue"><b>Generaing Random Dataset and spliting train test<b></h1>

```python
np.random.seed(42)
n_rows = 1000

df = pd.DataFrame({
    "Col-1":np.random.choice(["A11","A12","A13","A14","A15","A16","A17","A11","A12"],n_rows),
    "Col-2":np.random.choice(["S","A","I","S","A","I","A","V","S"], n_rows),
    "Target":np.random.choice([100, 200, 400, 200, 500, 600, 500, 400, 800],n_rows),
    })
```

```python
x = df.drop(columns="Target")
y = df.Target

Xtrain, Xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

<h1 style="color:lightgreen"><b>1. LabelEncoder</b></h1>

### **Working-**
        - selects unique columns
        - sort by acessding order
        - assigns values from 0
        - stores values in dictonary
ex:
        {
                "red":2,
                "c1":1,
                "c2":2,
        }



>### **Note - use only 1D array(Single column) | Error if more than 1 column**

```python
Lable = LabelEncoder()
updated_Xtrain = Lable.fit_transform(Xtrain["Col-1"])
updated_Xtrain

# Lable.transform(Xtrain["Col-2"])    Error of unseen data
```

<h1 style="color:lightgreen"><b>2. OneHotEncoder</b></h1>

### **Working-**
        - it encodes binary values
        - selects unique columns
        - sort by acessding order
        - creates columns based on unique values
        
        

><h3 style="color:orange"><b>Note - use only 2D array(more than 1 column) | Error if less than 2 column</b></h3>

<h3 style="color:#00FFFF">Important</h3>

        - it will return in the form of matrix
        - it will create more columns (columns of unique values)

---
<h3 style="color:#00FFFF"><pre><b>     Method - 1</b></pre></h3>

            - OneHot = OneHotEncoder()   # default sparse_output=True
            - in this method we  need to convert martix to read values


```python
OneHot = OneHotEncoder()


updated_Xtrain = OneHot.fit_transform(Xtrain)

# toarray() => this will return values
values = updated_Xtrain.toarray()
values
```

```python
# this will give column names(asc order)
colums = OneHot.get_feature_names_out()
colums
```

```python
# turning raw matrix into dataframe
pd.DataFrame(values, columns=colums)
```

---
<h3 style="color:#00FFFF"><pre><b>     Method - 2</b></pre></h3>

            - OneHot = OneHotEncoder(sparse_output=False)
            - in this method we dont need to convert to array
            - directly we can get array
            - use this method for projects and further

```python
# Default value for handle_unknown is error change to ignore it will  
OneHot = OneHotEncoder(sparse_output=False, handle_unknown="ignore")

updated_Xtrain = OneHot.fit_transform(Xtrain)
updated_Xtrain
```

```python

OneHot.transform([["X11","M"]])
```

```python
pd.DataFrame(updated_Xtrain,columns=colums)
```

<h1 style="color:lightgreen"><b>3. Ordinal Encoder</b></h1>


## **Working**
    - works for 1 column also
    - it will make a dictory and assing values accordingly starting from 0
    - works like lable encoder when order not specified
    - selects unique values and arranges as given order 
    - assigns values from 0

><h3 style="color:orange"><b>Note - it works for 1 column also and it gives random values if order not passed(Lable encoder)</b></h3>


## **Import**
    -from sklearn.preprocessing import OrdinalEncoder

## **unknown values**
    - it dont have handle_unknown="ignore"
    - we use handle_unknown="use_encoded_value", unknown_value=-1 while creating obj
    - here we assign -1 to unknown not 0 because the values assigment starts from 0 to inf
    - if we use 0 or any other +ve values it will mix the values and will get wrong values

---
<h3 style="color:#00FFFF"><pre><b>     Method - 1</b></pre></h3>


```python
Ord = OrdinalEncoder() # creating Obj
```

```python
updated_Xtrain = Ord.fit_transform(Xtrain) # only 2-D array
updated_Xtrain
```

```python
updated_Xtest = Ord.transform(Xtest) # only 2-D ande error when used 1D
updated_Xtest
```

<h1 style="color:lightgreen"><b>4. Target Encoder</b></h1>




### **Defination-** 
        - it is a encoding pre-processing technique which encodes the values based on target column, 
        - if the target column is continuse values then it takes avg() 
        - if the target column is categroical values then it takes majoritys

### **Importing Sytax-**
        from category_encoders import TargetEncodes
        encodes = TargetEncodes() # creating obj
        encodes.fit_transform(Xtrain, ytrain)  # Train the data
        encodes.transform(Xtest)               # Transform the data directly

### **Working-**

|col-1|col-2|Target|
|---|---|---|
|S|A1|100|
|I|A1|200|
|V|A2|50|
|V|A2|300|

        Step-1 :- Groups the unique values 
| | | | |     
|---|---|--|--|
|S|100|A1|100, 200|
|I|200|A2|50,30|
|V|50,300|


        Step 2 :- 
| | | | |     
|---|---|--|--|
|S|100|A1|150|
|I|200|A2|175|
|V|175|

        Step - 3:-
| Col-1|Col-2 |     
|---|---|
|100|150|
|200|150|
|175|175|
|175|175|


### **When to Use-**
        - when we have more unique values

><h3 style="color:orange"><b>Note - cols=[column names] => used to get specic columns to feed it and rest will remane same </b></h3>


TargetEncoder(cols=["Col-1"], handle_unknown="values")
              
>cols=["Col-1"] => only selected columns are transformed

>handle_unknown="values" => it will assign avg of all the values from target column

```python
Target = TargetEncoder(cols=["Col-1"], handle_unknown="values")
```

```python
updated_Xtrain = Target.fit_transform(Xtrain,ytrain)
updated_Xtrain
```

```python
updated_Xtest = Target.transform(Xtest)
updated_Xtest
```
